```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
```

```python
df = pd.read_csv('country_wise_latest.csv')
df = df.fillna(0)  # Fill missing values
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['Confirmed'], kde=True)
plt.title('Confirmed Cases Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df['Deaths'], kde=True, color='red')
plt.title('Deaths Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['Recovered'], kde=True, color='green')
plt.title('Recovered Cases Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.heatmap(df[['Confirmed', 'Deaths', 'Recovered', 'Active']].corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
scaler = MinMaxScaler()
normalized = scaler.fit_transform(df[['Confirmed', 'Deaths', 'Recovered', 'Active']])
df_normalized = pd.DataFrame(normalized, columns=['Confirmed', 'Deaths', 'Recovered', 'Active'])
```
